## Effects of People Characteristics on Their Heart Disease

## Investigation Overview
##### In this investigation, I wanted to look at the characteristics of People that could have heart disease the main focus will be on the age of people and chest pain type ,the slope of the peak exercise ST segment  and FastingBS

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("heart.csv")
data['HeartDisease'] = data['HeartDisease'].map({0:'normal',1:'heartDisease'}) 
data['FastingBS'] = data['FastingBS'].map({0:'>120',1:'other'}) 
```

## Dataset Overview
>Cardiovascular diseases (CVDs) are the number 1 cause of death globally, taking an estimated 17.9 million lives each year, which accounts for 31% of all deaths worldwide. Four out of 5CVD deaths are due to heart attacks and strokes, and one-third of these deaths occur prematurely in people under 70 years of age. Heart failure is a common event caused by CVDs and this dataset contains 11 features that can be used to predict a possible heart disease.

##  how many of these people in dataset have heart disease       

> here we see that most of people in this data have heart diseas ( '1' ) and they about 500 people

```python
sb.countplot(data['HeartDisease'])
plt.title("count of heart disease people");
```

## Distribution of age

> the age data like a normal distribution data 

```python
plt.hist(data['Age'])
plt.xlabel("age");
plt.ylabel("count");
plt.title("distribution of age");
```

## Age vs HeartDisease

> We see that most people who have heart disease who are over 55

```python
sb.histplot(data = data ,x = 'Age' , hue ='HeartDisease' , multiple="stack" , alpha = 0.6)
plt.title("the difference between disease and normal people by age");
```

## Gender vs Heart Disease

> as wee see that male have heart disease than female

```python
sb.histplot(data = data ,x = 'Sex' , hue ='HeartDisease',multiple="dodge" , shrink=.8)
plt.title("the difference between disease and normal people by gender");
```

# chest pain type vs Heart disease

> the most chest pain type that make people have heart disease is ASY type

```python
sb.histplot(data = data ,x = 'ChestPainType' , hue ='HeartDisease',multiple="dodge" , shrink=.8)
plt.title("the difference between disease and normal people by chest pain type");
```

## RestingECG vs heart disease

```python
sb.histplot(data = data ,x = 'RestingECG' , hue ='HeartDisease',multiple="dodge" , shrink=.8)
plt.title("the difference between disease and normal people by resting ECG");
```

## ST_Slope vs Heart disease

```python
sb.histplot(data = data ,x = 'ST_Slope' , hue ='HeartDisease',multiple="dodge" , shrink=.8)
plt.title("the difference between disease and normal people by st slope");
```

## FastingBS vs Heart Disease

```python
sb.histplot(data = data ,x = 'FastingBS' , hue ='HeartDisease',multiple="dodge" , shrink=.8)
plt.title("the difference between disease and normal people by fasting BS");
```

## Age and Heart Disease vs gender

```python
sb.histplot(data, x="Age", y="Sex", hue="HeartDisease")
plt.title("the difference between disease and normal people by age for each gender");
```

## RestingBP and heart disease vs age 

```python
sb.histplot(data = data , y = 'Age',x = 'RestingBP', hue = 'HeartDisease' )
plt.title("the difference between disease and normal people by resting BP for each age");
```
